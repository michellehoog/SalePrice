# Predicting House Prices

From the *House Prices: Advanced Regression Techniques* data set on Kaggle

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as sm
import missingno as mn
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
import pandas_profiling
```

## Getting the data

```python
#load train and test data
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv") #this data has no Sale Price, so need to split training into train and test sets to test model
train.shape
```

```python
#see variable data types
train.dtypes
```

## Explore missing values

```python
# look at amount of missing data
missingnr = train.isnull().sum().sort_values(ascending=False)
missingpercent = (missingnr/train.shape[0])*100

#show percentage missing data for variables that have missing values
print("The percentage missing data by feature is:")
round(missingpercent[missingpercent>0],2)
```

```python
#get an overview of the data and missingness
pandas_profiling.ProfileReport(train)
```

There are many missing values in features such as having a pool, a fence, and alley and a fireplace. Also a lot of missing data in size of lot frontage, which may be a good predictor of house prices. We can also visualise correlations between missing values using the missingno package.

```python
#heatmap of the amout of non-missing data
mn.heatmap(train);
```

Correlated missing data are related to specific features that a house may not have, e.g. all infomration on the garage is missing if the house does not have a garage; information about the basement is missing when the house does not have a basement. Houses that have missing data for the garage, are also more likely to have missing data for fireplace, fence and pool variables. We see the same information when we look at a dendogram of nullity correlations.

```python
#tree representing groupings of columns that have strong nullity correlations
mn.dendrogram(train);
```

## Explore relationship between variables

```python
#see correlation between variables
corrmat = train.corr()
corrmat
```

Several predictors show high multicollinearity e.g. basement SF and firstfloor SF, as well as garage area and garage cars. These variables should not be used inthe model at the same time.

```python
#heatmap of correlations
sns.heatmap(corrmat, square = True);
```

```python
# get pairplots for continuous variables
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt', 'Fireplaces']
sns.pairplot(train[cols], height = 2);
```

Sale Price is not normally distributed and its relationship with some of the variables is non-linear and shows heteroscedasticity. Log transform this variable before running the regression.

## Basic linear regression

```python
#separate predictor and outcome variables
x_train = train[['OverallQual', 'GrLivArea', 'GarageCars', 'FullBath', 'YearBuilt']]
y_train = train['SalePrice']

#scale predictor variables
scaler = preprocessing.StandardScaler()
x_train_s = scaler.fit_transform(x_train)   

#fit multiple regression with logged Sale Price
linear1 = LinearRegression()
linear1.fit(x_train_s, np.log(y_train))

#get predicted values
pred = linear1.predict(x_train_s)

#plot predicted against residual values
ax = sns.regplot(x=pred,y=np.log(y_train)-pred,lowess=True,line_kws={"color":"black"})
ax.set_title('Residual plot for log Price data.')
ax.set_xlabel('Predicted values')
ax.set_ylabel('Residuals')
plt.show();
```

Two clear outliers present

```python
print("Mean squared error: ",np.log(sum(np.power((y_train-pred),2))/x_train.shape[0]))
```
